```python
#creating the dataframe
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
drug_df=pd.read_csv("/content/drug200.csv")
drug_df
```

```python
#converting categorical data into numerical data using label encoding
#creating an instance and use fit_transform function to fit label
#here we convert column containing 'Sex' into numericals
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
drug_df['Sex']=le.fit_transform(drug_df['Sex'])
```

```python
#here we convert column containing 'BP' into numericals
le1=LabelEncoder()
drug_df['BP']=le1.fit_transform(drug_df['BP'])
```

```python
#here we convert column containing 'Cholesterol' into numericals
le2=LabelEncoder()
drug_df['Cholesterol']=le2.fit_transform(drug_df['Cholesterol'])
```

```python
drug_df
```

```python
#creating a variable called x which represents the input datas
x=drug_df.iloc[:,:-1].values
x
```

```python
#creating a variable called y which represents the output datas
y=drug_df.iloc[:,-1].values
y
```

```python
#splitting dataset into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30)
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
#converting th values into same range using normalization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
#create ML model
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=9)
knn.fit(x_train,y_train)
#predicting x_test
y_pred=knn.predict(x_test)
y_pred 
```

```python
#checking the perfromance of the model using confusion matrix,accuracy score,different errors
from sklearn.metrics import confusion_matrix,accuracy_score
mat=confusion_matrix(y_pred,y_test)
mat
```

```python
acc=accuracy_score(y_pred,y_test)
acc
```

```python
#to find the error rate
error_rate=[]
for i in range(1,40):
  knn=KNeighborsClassifier(n_neighbors=i)
  knn.fit(x_train,y_train)
  pred_i=knn.predict(x_test)
  error_rate.append(np.mean(pred_i!=y_test))
error_rate
```

```python
#plotting the error rate
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='b',marker='o',markerfacecolor='r',markersize=10)
plt.title("Error rate vs K value")
plt.xlabel('K')
plt.ylabel('Error rate')
```

```python
# A small value of k could lead to overfitting as well as a big value of k can lead to underfitting.
# Overfitting imply that the model is well on the training data but has poor performance when new data is coming.
# If k is selected to be too large, the model becomes too generalized and fails to accurately predict the data points in both train and test sets.
# thereby giving value of 9 to k
```
